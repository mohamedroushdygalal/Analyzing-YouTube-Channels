## Data Cleanning

```python
# replace NAN with 0 and change data type to integer 
df['video_views_rank'] = df['video_views_rank'].fillna(0).astype(int)
df['country_rank'] = df['country_rank'].fillna(0).astype(int)
df['channel_type_rank'] = df['channel_type_rank'].fillna(0).astype(int)
```

```python
df[['video_views_rank','country_rank','channel_type_rank']].info()
```

```python
# rename Gross tertiary education enrollment (%) column to enrollment_percentage
df.rename(columns={'Gross tertiary education enrollment (%)': 'enrollment_percentage'}, inplace=True)
df[["enrollment_percentage"]]
```

```python
# rename youtuber column to channel_name
df.rename(columns={'Youtuber': 'channel_name'}, inplace=True)
df[["channel_name"]]
```

```python
# dropping of duplicates
df.drop_duplicates(inplace=True)
```

```python
# delete row which have created year = 1970
df = df.drop(df[df['created_year'] == 1970].index)
```

```python
Data Analysis By Answering Some Questions 
```

```python
 # What are the ten YouTube channels with the highest number of subscribers?
top_five_channels = df[["channel_name","subscribers"]].head(10)
plt.figure(figsize=(10, 6))
plt.bar(top_five_channels["channel_name"],top_five_channels["subscribers"],color='skyblue',width=0.4)
plt.xlabel('channel_name')
plt.ylabel('subscribers')
plt.title('Top Ten Channels by Subscribers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
 # what is the total video views for each category?
total_video_views = df[['category', 'video views']].dropna(subset=['category'])
plt.figure(figsize=(10, 5))
plt.barh(total_video_views["category"],total_video_views["video views"],color='skyblue', height=0.6)
plt.xlabel('video views')
plt.ylabel('category')
plt.title('Total Video Views For Each Category')
plt.tight_layout()
plt.show()
```

```python
# What are the five most frequently occurring channel types among YouTube channels?
top5_channel_type= df['category'].value_counts().head()
plt.figure(figsize=(10, 4))
plt.pie(top5_channel_type, labels=top5_channel_type.index, autopct='%1.1f%%')
plt.title('Top 5 Most Common YouTube Channel Type')
plt.show()
```

```python
# Is there a correlation between the number of subscribers and video views?
correlation = df['subscribers'].corr(df['video views'])
print("Correlation between subscribers and video views:", correlation)
```

The Correlation coefficient is close to 1, it indicates a strong positive correlation, meaning as the number of subscribers increases, the number of video views also increases.

```python
# How does the total number of video views relate to the number of subscribers for each channel type?
# Group by 'channel_type' and calculate the sum of 'video views' and 'subscribers'
channel_type_stats = df.groupby('channel_type').agg({'video views': 'sum', 'subscribers': 'sum'})

# Normalize the data
normalized_video_views = (channel_type_stats['video views'] - channel_type_stats['video views'].min()) / (channel_type_stats['video views'].max() - channel_type_stats['video views'].min())
normalized_subscribers = (channel_type_stats['subscribers'] - channel_type_stats['subscribers'].min()) / (channel_type_stats['subscribers'].max() - channel_type_stats['subscribers'].min())

plt.figure(figsize=(10, 5))
plt.plot(channel_type_stats.index, normalized_video_views, marker='o', label='Sum Video Views')
plt.plot(channel_type_stats.index, normalized_subscribers, marker='o', label='Sum Subscribers')
plt.xlabel('Channel Type')
plt.ylabel('Normalized Value')
plt.title('Video Views and Subscribers by Channel Type')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Which countries have the highest total video views?
# Exclude rows with NaN values 
country_video_views = df[["Country", "video views"]].dropna()
plt.figure(figsize=(10, 5))
plt.bar(country_video_views["Country"], country_video_views["video views"], color='skyblue', width=0.7)
plt.xlabel('country')
plt.ylabel('video views')
plt.title('Total Video Views For Each Country')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#How can we create a visual representation to show the number of channels created over the years?
years = df['created_year'].dropna()
plt.figure(figsize=(12, 5))
plt.hist(years, bins=range(int(min(years)), int(max(years)) + 1, 1), edgecolor='black', alpha=0.7)
plt.xlabel('Year of Channel Creation')
plt.ylabel('Frequency')
plt.title('YouTube Channel Creation Years Histogram')
plt.xticks(range(int(min(years)), int(max(years)) + 1)) 
plt.xticks(rotation=45) 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#What is the relationship between video views and gross tertiary education enrollment 
#across different countries? 

# Convert 'video views' column to numeric, coercing errors to NaN
df['video views'] = pd.to_numeric(df['video views'], errors='coerce')

# Convert 'enrollment_percentage' column to numeric, coercing errors to NaN
df['enrollment_percentage'] = pd.to_numeric(df['enrollment_percentage'], errors='coerce')

# Drop rows with NaN values in 'video views' column
df_clean = df.dropna(subset=['video views'])

# Drop rows with non-numeric values in 'enrollment_percentage' column
df_clean = df_clean[pd.to_numeric(df_clean['enrollment_percentage'], errors='coerce').notnull()]

# Group by country and calculate the mean of 'video views' and 'enrollment_percentage'
grouped_df = df_clean.groupby('Country').agg({'video views': 'mean', 'enrollment_percentage': 'mean'}).reset_index()

# Plotting
plt.figure(figsize=(10, 5))
plt.scatter(grouped_df['video views'], grouped_df['enrollment_percentage'], color='skyblue', edgecolors='black', alpha=0.7)

# Add country names as annotations
for i, (views, enrollment, country) in enumerate(zip(grouped_df['video views'], grouped_df['enrollment_percentage'], grouped_df['Country'])):
    plt.text(views, enrollment, country, fontsize=7, ha='right', va='bottom')

# Add labels and title
plt.xlabel('Video Views')
plt.ylabel('Gross Tertiary Education Enrollment (%)')
plt.title('Video Views vs Gross Tertiary Education Enrollment by Country (Grouped)')

# Remove grid lines
plt.grid(False)

plt.tight_layout()
plt.show()
```
